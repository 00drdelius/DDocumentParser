```python
import requests
import rich

api_key="dataset-SwjM8mcKsTxXK3yD7ypvQ6cC"
base_url="http://19.119.245.93/v1/"
headers=dict(Authorization=f"Bearer {api_key}")
```

```python
with requests.get(base_url+"datasets",headers=headers,params=dict(page=1,limit=100)) as resp:
    rich.print(resp.json())
    data=resp.json()

chosen_dataset_id=[ item for item in data['data'] if item['name']=='审计局政策'][0]['id']
```

```python
payload=dict(
    query="中山市审计局财务管理制度第4章第15条复述一遍",
    retrieval_model=dict(
        search_method="semantic_search",
        reranking_enable=False,
        top_k=150,
        score_threshold_enabled=False,
    )
)
with requests.post(base_url+"datasets/{dataset_id}/retrieve".format(dataset_id=chosen_dataset_id),
                   headers=headers,
                   json=payload) as resp:
    resp.raise_for_status()
    data=resp.json()
segments=[ i['segment']['content'] for i in data['records']]
rich.print(segments)
```
